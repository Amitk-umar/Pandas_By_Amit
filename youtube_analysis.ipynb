# Which categories have the most videos.

```python
# Count number of videos in each category
category_counts = data['category'].value_counts()

category_counts
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
category_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("YouTube Video Count by Category")
plt.xlabel("Category")
plt.ylabel("Number of Videos")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
data['comments'].value_counts().head(5)
```

```python
comments = data.groupby('comments')['title'].value_counts().reset_index().sort_values(by='comments', ascending=False).head(5)
```

```python
comments
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(comments['title'], comments['comments'], color='teal')
plt.xlabel("Number of Comments")
plt.ylabel("Video Title")
plt.title("Top 5 Most Commented Videos")
plt.gca().invert_yaxis()  # Highest at the top
plt.show()
```
