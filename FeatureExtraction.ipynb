# Q. Make a new column for episodes count


```python
#make a new column for episode count
df.loc[2]['Title']
```

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if check == True:
            data = data + i
        if  i == '(':
            check = True
            
        if i == ')':
            check = False
            return data
```

```python
df['Title']
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)
```

```python
df['Episodes'] = df['Episodes'].str.replace('eps)', '',)
df
```

```python
def extraction_timestamp(txt):
    check = False
    data = ""
    for i in range(len(txt)):
     if txt[i] == ')':
        for j in range(i+1,i+20):
            data += txt[j]
        return data   
```

```python
df['Time Stamp'] = df['Title'].apply(extraction_timestamp)  # Apply the extraction function to the 'Title' column
```

# making a column for Months

```python
# Function to calculate number of months
def count_months(time_range):
    start_str, end_str = time_range.split(" - ")
    start = pd.to_datetime(start_str, format="%b %Y")
    end = pd.to_datetime(end_str, format="%b %Y")
    
    # Calculate difference in months (inclusive)
    months = (end.year - start.year) * 12 + (end.month - start.month) + 1
    return months
# Example: "Apr 2009 - Jul 2010"
# Years difference: 2010 - 2009 = 1 → 12 months
# Month difference: 7 - 4 = 3 months
# Total: 12 + 3 + 1 = 16 months
# Apply function
df["Months"] = df["Time Stamp"].apply(count_months)
df.head()
```
